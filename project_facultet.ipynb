```python
import requests
import json

url = "https://api.coingecko.com/api/v3/simple/price"
params = {
    "ids": "ethereum",
    "vs_currencies": "usd"
}

response = requests.get(url, params=params)
response.raise_for_status()

data = response.json()

with open("ethereum_price.json", "w") as file:
    json.dump(data, file, indent=4)

print(data)
```
